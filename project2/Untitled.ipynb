```python
import pandas as pd
import torch
import chardet
from transformers import RobertaTokenizer, RobertaForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset

# Step 1: Detect encoding and read dataset
def detect_encoding(file_path):
    """Detects the file encoding to prevent UnicodeDecodeError."""
    with open(file_path, "rb") as f:
        result = chardet.detect(f.read())
    return result["encoding"]

def load_dataset(file_path):
    """Loads CSV or Excel file with proper encoding."""
    try:
        if file_path.endswith(".xlsx"):  # If it's an Excel file
            return pd.read_excel(file_path)
        elif file_path.endswith(".csv"):  # If it's a CSV file
            encoding = detect_encoding(file_path)
            return pd.read_csv(file_path, encoding=encoding)
        else:
            raise ValueError("Unsupported file format! Use .csv or .xlsx")
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None

# Load labeled and unlabeled datasets
labeled_data = load_dataset("/content/labeled_data.csv")  # Your manually labeled dataset
unlabeled_data = load_dataset("/content/unlabeled_data.csv")  # Your dataset to label

if labeled_data is None or unlabeled_data is None:
    print("Error: Unable to load datasets. Check file paths and formats.")
    exit()

# Step 2: Load RoBERTa tokenizer and model
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
model = RobertaForSequenceClassification.from_pretrained("roberta-base", num_labels=2)  # Binary classification

# Step 3: Create a custom PyTorch dataset
class EthicalDataset(Dataset):
    def __init__(self, texts, labels=None):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        encoding = tokenizer(self.texts[idx], padding="max_length", truncation=True, max_length=128, return_tensors="pt")
        item = {key: val.squeeze(0) for key, val in encoding.items()}
        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

# Step 4: Prepare datasets
train_texts = labeled_data["Comment"].astype(str).tolist()  # Ensure it's a string
train_labels = labeled_data["Label"].tolist()

train_dataset = EthicalDataset(train_texts, train_labels)

# Step 5: Set up Trainer for fine-tuning
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir="./logs",
    per_device_train_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
)

# Step 6: Train the model
trainer.train()

# Step 7: Predict on Unlabeled Data
unlabeled_texts = unlabeled_data["Comment"].astype(str).tolist()
unlabeled_dataset = EthicalDataset(unlabeled_texts)

predictions = trainer.predict(unlabeled_dataset)
pred_labels = torch.argmax(torch.tensor(predictions.predictions), dim=1).tolist()

# Step 8: Save results
unlabeled_data["Predicted_Label"] = pred_labels
unlabeled_data.to_csv("labeled_unlabeled_data.csv", index=False)

print("✅ Predictions saved to labeled_unlabeled_data.csv!")
```
